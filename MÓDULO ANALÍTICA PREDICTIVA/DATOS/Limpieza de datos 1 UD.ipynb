#PRIMER SEMESTRE INDUSTRIAL

```python
industrial = industrial.apply(pd.to_numeric)
```

```python
# Lista de todas las materias
materias = [col.split('_')[1] for col in industrial.columns if col.startswith('NOTA_')]

# Lista para almacenar los promedios por estudiante
promedios = []

# Iterar sobre cada estudiante
for index, row in industrial.iterrows():
    notas_estudiante = []
    creditos_estudiante = []
    # Iterar sobre las materias
    for materia in materias:
        veces_column = f'VECES_{materia}'
        nota_column = f'NOTA_{materia}'
        creditos_column = f'CREDITOS_{materia}'
        # Verificar si el estudiante cursÃ³ la materia al menos una vez
        if row[veces_column] > 0:
            # Agregar la nota y los crÃ©ditos a las listas correspondientes
            notas_estudiante.append(row[nota_column])
            creditos_estudiante.append(row[creditos_column])
    # Calcular el promedio ponderado para el estudiante actual
    if len(notas_estudiante) > 0:
        promedio_estudiante = np.average(notas_estudiante, weights=creditos_estudiante)
    else:
        # Si el estudiante no cursÃ³ ninguna materia, asignar promedio cero
        promedio_estudiante = 0
    # Agregar el promedio del estudiante a la lista de promedios
    promedios.append(promedio_estudiante)

# Agregar la lista de promedios al DataFrame
industrial['Promedio'] = promedios

industrial
```
